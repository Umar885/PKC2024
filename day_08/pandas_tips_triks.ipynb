# 16 - Splitting a string into multiple columns

```python
# import the libraries 
# create the dataframe
import pandas as pd
df = pd.DataFrame({'Name': ["Umar Mehmood", "Usman Khan", "Ali Murtaza", "Babar Awan", "Khalid Hussain"], 'Location': ["Rahim Yar Khan, Pakistan", "Lahore, Pakistan", "Karachi, Pakistan", "Islamabad, Pakistan", "Lahore, Pakistan"]})
df
```

```python
df[["first_name", "last_name"]] = df["Name"].str.split(" ", expand = True)
df
```

```python
#split the location
df[["city", "country"]] = df["Location"].str.split(",", expand = True)
df
```

```python
# Refine the data manipulation
df = df[["first_name", "last_name", "city", "country"]]
df
```

# 23 - Reshaping  a dataframe

```python
false = pd.DataFrame([['12345', 100, 200, 300], ['34567', 200, 300, 400], ['56789', 300, 400, 500]],
                     columns=['zip', 'factory', 'werehouse', 'retail'])
false.head()
```

```python
false2 = pd.DataFrame([[1, "12345", "factory"], [2, "34567", "werehouse"], [3, "56789", "retail"]], 
                      columns=['id', 'zip', 'location'])
false2.head()
```

```python
fasla = false.melt(id_vars=['zip'], value_vars=['factory', 'werehouse', 'retail'])
fasla
```

```python
import seaborn as sns
# create a bar plot
sns.barplot(x='zip', y='value', hue='variable', data=fasla)
```
